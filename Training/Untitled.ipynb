```python
import pandas as pd
```

```python
def read_csv_from():
    smart=pd.read_csv("smartphones.csv")
    return smart
read_csv_from()
```

```python
read_csv_from().info()
```

```python
read_csv_from().describe().T
```

```python
def unwanted_columns():
    smart=read_csv_from()
    unwanted_columns=smart.drop(columns=['has_5g','has_nfc','has_ir_blaster','model'],axis=1,inplace=True)
    return smart
```

```python
unwanted_columns()
```

```python
unwanted_columns().columns
```

```python
def check_null_values():
    smart=unwanted_columns()
    smart.dropna(inplace=True)
    return smart
```

```python
check_null_values().isnull().sum()
```

```python
def find_duplicate():
    smart=check_null_values()
    smart=smart.drop_duplicates(keep=False)
    return smart
```

```python
find_duplicate().columns
```

```python
find_duplicate()
```

```python
pd.options.display.max_columns=50
pd.options.display.max_rows=999
```

```python
def convert():
    smart=find_duplicate()
    smart['rating']=smart['rating'].astype('int')
    smart['num_cores']=smart['num_cores'].astype('int')
    smart['processor_speed']=smart['processor_speed'].astype('int')
    smart['battery_capacity']=smart['battery_capacity'].astype('int')
    smart['fast_charging_available']=smart['fast_charging_available'].astype('int')
    smart['fast_charging']=smart['fast_charging'].astype('int')
    smart['num_front_cameras']=smart['num_front_cameras'].astype('int')
    smart['primary_camera_rear']=smart['primary_camera_rear'].astype('int')
    smart['primary_camera_front']=smart['primary_camera_front'].astype('int')
    smart['extended_upto']=smart['extended_upto'].astype('int')
    return smart
```

```python
convert()
```

```python
def to_csv():
    dataframe=convert()
    dataframe.to_csv("cleandata.csv")
    return dataframe
```

```python
to_csv()
```
